```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Top 10 kdramas based on the rankings.

```python
df_dramas = pd.read_csv('kdrama.csv')
df_dramas.head(10)
```

Checking the statistics of the given data.

```python
df_dramas.describe()
```

As you can see from the above data that only korean dramas that were released till 2022 were recorded and we can also confirm that this is the data of top 250 as min rating is 8.30.

Display kdramas which were released after 2020.

```python
after2020= df_dramas[df_dramas['Year of release']>= 2020]
after2020.head()
```

Now, lets display the top 10 dramas which were released after 2020

```python
after2020.head(10)
```

Let's see what is the average rating of a director and rank them.

```python
directors = df_dramas.groupby('Director')['Rating'].mean().sort_values(ascending=False)
directors.head(10)
```

Display 10 best movies which have less than 60 mins of runtime

```python
lessthan60 = df_dramas[df_dramas['Duration (in mins)']<=60]['Name']
lessthan60.head(10)
```

Display kdramas with longest runtime

```python
top10_length = df_dramas.nlargest(10, 'Duration (in mins)') [['Name', 'Duration (in mins)']]\
.set_index('Name')
top10_length.head(10)
```

Now lets visualize this data

```python
sns.barplot(x='Duration (in mins)', y=top10_length.index, data= top10_length)
```

Display the number of Kdramas(top 250) per year.

```python
after2011= df_dramas[df_dramas['Year of release']>= 2011]

after2011['Year of release'].value_counts()
```

```python
sns.countplot(x='Year of release',data= after2011)
plt.title("Number of top kdramas per year")
```

Display top 10 best movies and its directors in graph

```python
top10_len  = df_dramas.nlargest(10, 'Rating')[['Name','Rating','Director']]\
.set_index('Name')
top10_len
```

Find average rating of top kdramas in different years

```python
after2011.groupby('Year of release')['Rating'].mean().sort_values(ascending=False)
```
